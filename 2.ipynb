```python
import numpy as np
shape = (3, 3, 6)
three_D_array = np.random.randint(0, 21, shape)
print(three_D_array)
print('배열의 shape:', three_D_array.shape)
```

```python
import numpy as np
shape = (5, 6)
two_D_array = np.random.randint(0, 101, shape)
print(two_D_array)
print('배열의 shape:', two_D_array.shape)
```

```python
import numpy as np
temp = np.arange(48)
print('Original 1D Array temp shape:', temp.shape)

# 生成三维数组
three_D_array = temp.reshape(4, 3, 4)
print('\nReshaped 3D Array (4, 3, 4):')
print(three_D_array)
print('Shape of 3D array:', three_D_array.shape)

# 生成二维数组
two_D_array = temp.reshape(6, 8)
print('\nReshaped 2D Array (6, 8):')
print(two_D_array)
print('Shape of 2D array:', two_D_array.shape)
```
